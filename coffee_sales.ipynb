## Calculate sum(transaction_qty) of with each product category

```python
df.pivot_table(index='product_category',values='transaction_qty',aggfunc='sum')
```

## To find sum of transaction quantity with store_loation

```python
df.groupby(by='store_location')[['transaction_qty']].sum()
```

```python
result_df=df.groupby(by='store_location')[['transaction_qty']].sum()
```

```python
import matplotlib.pyplot as plt
colors = ['red', 'green','blue']
plt.figure(figsize=(8,5))
plt.xlabel("Store Location",fontsize=15,color='red')
plt.ylabel("Transaction Quantity",fontsize=15,color='red')
plt.bar(result_df.index,result_df['transaction_qty'],color=colors,width=0.5)
plt.show()
```

## Calculate the  total Revenue generated  by Store location

```python
df['revenue']=df['transaction_qty']*df['unit_price']
```

```python
result_df=df.groupby(by='product_category')[['revenue']].sum()
```

```python
color=['green', 'orange', 'blue','purple','red','yellow','orangered','violet','black']
plt.xlabel("Total Revenue",color='red',fontsize=20)
plt.ylabel("Store Location",color='red',fontsize=20)
plt.barh(result_df.index,result_df['revenue'],color=color)
plt.show()
```
